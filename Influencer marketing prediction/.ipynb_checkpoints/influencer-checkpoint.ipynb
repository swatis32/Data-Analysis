### 2. Machine Learning

```python
X_data=data.drop(labels=['Sales'], axis=1)
X_data.head()

Y_data=data['Sales']
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
X_data['Influencer']=encoder.fit_transform(X_data.loc[:,'Influencer'])

dict(zip(encoder.classes_, encoder.fit_transform(encoder.classes_)))
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_data,Y_data, random_state=1, test_size=0.2 )
```

```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

### 2.1 Cross validate and find the best Models

### 2.1.1 Linear Regression

```python
lr=LinearRegression()

scores=cross_validate(lr,X=X_train,y=y_train, cv=7,scoring='r2', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])

#lr.fit(X_train,y_train)
#lr.coef_
```

### 2.1.2 Random Forest Regressor

```python
rf=RandomForestRegressor(random_state=1)

scores=cross_validate(rf,X=X_train,y=y_train, cv=7,scoring='r2', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])
```

### 2.1.3 Gradient Boosting Regressor

```python
gb=GradientBoostingRegressor(random_state=1)

scores=cross_validate(gb,X=X_train,y=y_train, cv=7,scoring='r2', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])
```

### 2.1.4 XGB Regressor

```python
xg=XGBRegressor(random_state=1)

scores=cross_validate(xg,X=X_train,y=y_train, cv=7,scoring='r2', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])
```

### 2.2 Hyper Tuning parameters

### 2.2.1 Random Forest Regressor

```python
from sklearn.model_selection import GridSearchCV

rf= RandomForestRegressor(random_state=1)

parameters={'max_depth': [2,3,4,5],
           'n_estimators': [50,100,200]}

grid1=GridSearchCV(rf, parameters, cv=7, scoring='r2', return_train_score=True)
grid1.fit(X_train,y_train)
grid1.cv_results_["mean_train_score"][grid1.best_index_]
grid1.cv_results_["mean_test_score"][grid1.best_index_]



```

```python
rf_final=grid1.best_estimator_
y_pred=rf_final.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2_score(y_test,y_pred)
#mean_absolute_error(y_test,y_pred)
#mean_squared_error(y_test,y_pred)
```

### 2.2.2 Gradient Boosting Regressor

```python
gb= GradientBoostingRegressor(random_state=1, max_depth=5)

parameters={'learning_rate': [.01, .1, 1, 10],
           'n_estimators': [50,100,200]}

grid2=GridSearchCV(gb, parameters, cv=7, scoring='r2', return_train_score=True)
grid2.fit(X_train,y_train)
grid2.cv_results_["mean_train_score"][grid2.best_index_]
grid2.cv_results_["mean_test_score"][grid2.best_index_]

```

```python
gb_final=grid1.best_estimator_
y_pred=rf_final.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2_score(y_test,y_pred)
#mean_absolute_error(y_test,y_pred)
#mean_squared_error(y_test,y_pred)
```
