### 2. Machine Learning Models

### 2.1 Create X_data and y_data

```python
X_data=data.drop(labels=['enrollee_id','city','experience','last_new_job','target'], axis=1)
y_data=data['target']
X_data.head()
```

### 2.2 App apply standard scaler and label encoding

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
X_data['gender']=encoder.fit_transform(X_data.loc[:,'gender']) #note all 3 ways are correct
X_data['relevent_experience']=encoder.fit_transform(X_data['relevent_experience']) #note all 3 ways are correct
X_data['enrolled_university']=encoder.fit_transform(X_data[['enrolled_university']]) #note all 3 ways are correct
X_data['education_level']=encoder.fit_transform(X_data.loc[:,'education_level']) 
X_data['major_discipline']=encoder.fit_transform(X_data.loc[:,'major_discipline']) 
X_data['company_size']=encoder.fit_transform(X_data.loc[:,'company_size']) 
X_data['company_type']=encoder.fit_transform(X_data.loc[:,'company_type']) 

dict(zip(encoder.classes_,encoder.fit_transform(encoder.classes_)))
```

### 2.3 Test train data split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_data,y_data, random_state=1, test_size=0.2)

```

### 2.4 Compare models 

### 2.4.1 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate

#y_train.value_counts()[0]/y_train.value_counts()[1] # class__weight

lr=LogisticRegression(random_state=1, class_weight={0:1,1:5})

scores=cross_validate(lr,X=X_train, y=y_train, cv=7,scoring='accuracy', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])


```

### 2.4.2 Random Forest Classifier

```python
rf=RandomForestClassifier(random_state=1, class_weight={0:1,1:5})

scores=cross_validate(rf,X=X_train, y=y_train, cv=7,scoring='accuracy', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])
```

### 2.4.3 XGB Classifier: it is gradient boosting with regularization

```python
# https://machinelearningmastery.com/xgboost-for-imbalanced-classification/
# above contains the documentation of scale_pos_weight
xgb_c=XGBClassifier(random_state=1, scale_pos_weight=5)

scores=cross_validate(xgb_c,X=X_train, y=y_train, cv=7,scoring='accuracy', return_train_score=True)
np.mean(scores["train_score"])
np.mean(scores["test_score"])
```

### 2.5 Tune parameters for RandomForestClassifier and XGBClassifier

### 2.5.1 RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf=RandomForestClassifier(random_state=1, class_weight={0:1,1:5})
parameters={'n_estimators': [50,100,200],
            'max_depth': [2,3,4,5]}

grid1=GridSearchCV(rf,parameters,cv=7, scoring='accuracy',return_train_score=True)
grid1.fit(X_train,y_train)

grid1.cv_results_["mean_train_score"][grid1.best_index_]
grid1.cv_results_["mean_test_score"][grid1.best_index_]
grid1.best_params_

```

```python
rf_final=grid1.best_estimator_
y_pred=rf_final.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred, target_names=['No Job Switch', 'Job Switch']))
```

```python
from sklearn.metrics import confusion_matrix

ans=confusion_matrix(y_test,y_pred)
sns.heatmap(ans, annot=True, fmt='d', cbar=False, xticklabels=['No Job Switch', 'Job Switch'],yticklabels=['No Job Switch', 'Job Switch'])
```

```python
plt.bar([x for x in range(len(rf_final.feature_importances_))],rf_final.feature_importances_)

#X_train.head()
```

### 2.5.2 XGBClassifier

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

xg=XGBClassifier(random_state=1, scale_pos_weight=5)

parameters={'n_estimators':[50,100,250],
           'learning_rate':[.01,.1,1,10]}

grid2=GridSearchCV(xg, parameters, cv=7, scoring='accuracy', return_train_score=True)
grid2.fit(X_train,y_train)

grid2.best_params_
grid2.cv_results_["mean_train_score"][grid2.best_index_]
grid2.cv_results_["mean_test_score"][grid2.best_index_]



```

```python
xg_final=grid2.best_estimator_
y_pred=xg_final.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,y_pred)
```

```python
from  sklearn.metrics import classification_report

print(classification_report(y_test,y_pred, target_names=['No job switch','job switch']))
```

```python
from  sklearn.metrics import confusion_matrix

ans=confusion_matrix(y_test,y_pred)
                      
sns.heatmap(ans, annot=True, cbar=False, fmt='d', xticklabels=['No job switch','job switch'], yticklabels=['No job switch','job switch'])
```
